```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("economicdata2022-2022.csv")

# In the accompanying text, report on max, min, mean, median.
print(df.info())
print(df.describe())

# Plot 1-dimensional histogram for numerical variables.
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

for col in numerical_cols:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Histogram for {col}')
    plt.show()


#No data in the dataset is categorical, so there is no need to plot a bar chart.
#Box plot for numerical variables including outliers.
plt.figure(figsize=(20, 10))
sns.boxplot(data=df)
plt.ylim(0, 300)
plt.title('Box Plot of Numerical Variables')
plt.xticks(rotation=45)
plt.show()

sns.pairplot(df[numerical_cols])
plt.show()

# Categorical vs Numerical (Color-coded Scatter Plot)
sns.scatterplot(data=df)
plt.title('Categorical vs Numerical Relationship')
plt.show()
```
